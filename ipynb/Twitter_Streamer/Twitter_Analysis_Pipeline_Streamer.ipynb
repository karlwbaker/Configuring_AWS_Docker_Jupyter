```python
"""Signature: collection_client.count_documents(filter, session=None, **kwargs)
Docstring:
Count the number of documents in this collection.
"""

# query Specifies selection filter using query operators. To return all documents in a collection, 
# omit this parameter or pass an empty document ({}).
# collection_client.count_documents(filter=$expr('karl'))

collection_client.count_documents(filter=({}))
```

```python
"""$exists
Syntax: { field: { $exists: <boolean> } }

EXAMPLE???  { document_name: { $in: true, 'karl' } }

When <boolean> is true, $exists matches the documents that contain the field, including documents where the field value is null. If <boolean> is false, the query returns only the documents that do not contain the field.

MongoDB $exists does not correspond to SQL operator exists. For SQL exists, refer to the $in operator."""

# collection_client.count_documents(filter=({ document_name: { $in: true, 'karl' } }))
collection_client.count_documents(filter={ document_name: { $in: true, 'karl' } })
```
